获取当前进程编号Process ID = PID

​	multiprocessing.current_process().pid

获取当前进程的父进程的 ID    parent PID = PPID

​	os.getppid()

>   获取系统中进程的信息            `ps aux | grep .py`

>    终止指定 PID 的进程           os.kill(进程PID, 9) # 9号信号是绝对退出的信号

```python
import multiprocessing
import os
import time


def sing(sname, singer):
    """子进程１运行当前函数"""
    for i in range(10000):
        print("%s 唱 %s....　PID=%s PPID=%s" % (singer, sname, os.getpid(),os.getppid()))
        # alt+enter 快速导入模块
        time.sleep(1)  # 休息１秒中

def dance(dname, dancer):
    """子进程2运行当前函数"""
    for i in range(10000):
        # 获取当前进程的PID os.getpid() 和 multiprocessing.current_process().pid)
        print("%s 跳 %s.... PID=%s" % (dancer, dname, multiprocessing.current_process().pid))
        time.sleep(1)  # 休息１秒中

def main():

    sing_pro = multiprocessing.Process(target=sing,
                                       args=('月亮之上', '玲花'))
    # 2 一定要调用这个对象.start()　启动子进程创建和运行
    sing_pro.start()

    dance_pro = multiprocessing.Process(target=dance,
                                        kwargs={'dname':'太空漫步', 'dancer':"迈克"},
                                        name='dancexxxxx')
    dance_pro.start()

    # time.sleep(3)
    # sing_pro.terminate()  # 不等子进程正常结束　　强制终止子进程
    # dance_pro.terminate()

    # 主进程　需要等待子进程运行完成  直到子进程退出　主进程才会继续往下执行代码
    sing_pro.join()
    dance_pro.join()

    print("感谢诛仙天雷滚滚")
    print("PID=%s" % os.getpid())


if __name__ == '__main__':
    main()
```

```python
import multiprocessing
import os
import time


def func():
    for i in range(5):
        print("PID = %s" % os.getpid())
        time.sleep(1)


if __name__ == '__main__':
    # 创建很多进程  给每个进程打印PID
    for i in range(10):
        p = multiprocessing.Process(target=func)
        p.start()

    # 多进程执行顺序和系统调度策略 随机的
    # 启动这个代码 主进程的作用启动10子进程 然后就没有代码可以执行了
    # ps aux| grep python3 发现还有１１个进程　说明主进程运行完代码之后没有退出而是等待所有子进程一起退出
```



### 3.5 daemon 进程意义

​	默认情况下 主进程执行完代码会等待所有子进程一起退出

​	如果我们把子进程设置 daemon，导致主进程退出时带领所有daemon子进程销毁, 非 daemon 子进程还在继续 run 执行。

​	设置  进程对象.daemon = True

```python
import multiprocessing
import os
import time


def func():
    for i in range(5):
        print("PID = %s" % os.getpid())
        time.sleep(1)

def func2():
    for i in range(5):
        print("PID = %s" % os.getpid())
        time.sleep(1)

if __name__ == '__main__':
    # 创建很多进程  给每个进程打印PID

    p = multiprocessing.Process(target=func)
    # 如果将子进程设置为daemon子进程　　主进程退出时候就会带领所有ｄａｅｍｏｎ子进程全部退出
    p.daemon = True  # 一定要放在启动之前
    p.start()

    p2 = multiprocessing.Process(target=func2, daemon=True)
    p2.start()

    # 多进程执行顺序和系统调度策略 随机的
    # 启动这个代码 主进程的作用启动10子进程 然后就没有代码可以执行了
    # ps aux| grep python3 发现还有１１个进程　说明主进程运行完代码之后没有退出而是等待所有子进程一起退出
```

```python
import threading
import time


def sing(sname, singer):
    """子进程１运行当前函数"""
    for i in range(10):
        print("%s 唱 %s...." % (singer, sname))
        time.sleep(1)  # 休息１秒中

def dance(dname, dancer):
    """子进程2运行当前函数"""
    for i in range(10):
        print("%s 跳 %s...." % (dancer, dname))
        time.sleep(1)  # 休息１秒中


def main():
    """从这里执行的代码都是在主线程中执行的"""
    # 1 创建两个子线程对象  一个线程唱歌  一个线程跳舞
    sing_thd = threading.Thread(target=sing, args=('凤凰传奇','荷塘月色'))
    dance_thd = threading.Thread(target=dance, kwargs={'dname': "篮球舞蹈", 'dancer':'我坤哥'})

    # 2 启动执行
    sing_thd.start()
    dance_thd.start()

    # 3 等待子线程运行完成
    sing_thd.join()
    dance_thd.join()
    print("哇哇哇  惊为天人啊")


if __name__ == '__main__':
    main()
```

    # 需求: 希望在主线程执行完代码之后　能够带领所有的子线程一起强制退出
    # 如果有10子线程 全部都是daemon线程　主线程退出所有线程全部强制退出
    # 如果有10子线程　　其中９个是daemon 其中１个是非daemon　主线程退出　任意子线程并不会强制退出;
    
    # daemon线程:    (当整个进程中剩下的全是daemon线程时　　整个进程中所有的线程　将会强制退出)
    # daemon子线程会　跟随最后最后一个daemon线程完成　一起强制退出


​    
    # daemon进程:    daemon子进程会　跟随主进程完成　一起强制退出
```python
import threading

import time


def func():
    for i in range(10):
        print("%s" % threading.current_thread().ident)  # 线程编号
        # 获取当前进程中所有的线程列表
        print(threading.enumerate())

        time.sleep(1)


def main():
    for i in range(5):
        # 1 创建5个线程  给每个线程一个编号
        thd = threading.Thread(target=func)
        thd.daemon = True

        # 2 启动每个线程
        thd.start()
    # 多线程运行顺序是无序的
    # 主线程运行完代码之后　不会立即退出　而是会等待所有的子线程一起退出
    print("主线程最后一行代码")
if __name__ == '__main__':
    main()
```

资源竞争问题

```python
import threading

g_number = 0

def number1():
    """数数线程　获取全局变量　＋１赋值给全局变量"""
    global g_number
    for i in range(1000000):
        g_number += 1


def number2(lock):
    """数数线程　获取全局变量　＋１赋值给全局变量"""
    global g_number
    for i in range(1000000):
        g_number += 1



if __name__ == '__main__':
    # 1 创建两个子线程　全局变量
    thd1 = threading.Thread(target=number1)
    thd2 = threading.Thread(target=number2)

    # 2 启动线程运行
    thd2.start()
    thd1.start()

    # 3 等待子线程运行完成　　打印全局变量
    thd1.join()
    thd2.join()

    print("最终获取全局变量的值是%s" % g_number)
```



资源竞争: 多个线程不加限制的同时修改共享的全局变量导致数据出错的现象

互斥锁特点: 保证任意时刻只有一个任务占有锁



互斥锁的使用

```
# 1 创建互斥锁
互斥锁对象 = threading.Lock()
# 2 申请互斥锁
互斥锁对象.acquire()
# 3 释放互斥锁
互斥锁对象.release()
```

```python
import threading

g_number = 0

def number1(lock):
    """数数线程　获取全局变量　＋１赋值给全局变量"""
    global g_number
    for i in range(1000000):
        # 1 申请锁
        lock.acquire()
        g_number += 1
        # 2 释放锁
        lock.release()

def number2(lock):
    """数数线程　获取全局变量　＋１赋值给全局变量"""
    global g_number
    for i in range(1000000):
        lock.acquire()
        g_number += 1
        lock.release()


if __name__ == '__main__':
    # 0 为两个线程同步操作创建一个互斥锁
    lock = threading.Lock()

    # 1 创建两个子线程　全局变量
    thd1 = threading.Thread(target=number1, args=(lock,))
    thd2 = threading.Thread(target=number2, args=(lock,))

    # 2 启动线程运行
    thd2.start()
    thd1.start()

    # 3 等待子线程运行完成　　打印全局变量
    thd1.join()
    thd2.join()

    print("最终获取全局变量的值是%s" % g_number)
```

​	优点: 保持同步解决资源竞争

​	缺点: 丧失多任务优势、申请和释放锁消耗系统资源

​			如果忘记释放锁   导致死锁问题

死锁问题:

​	记得在合适的位置释放锁

```python
def number2(lock):
    """数数线程　获取全局变量　＋１赋值给全局变量"""
    global g_number
    for i in range(1000000):
        # lock.acquire()
        with lock:
            g_number += 1
        # lock.release()
        
# with 能够自动申请和释放互斥锁  减少死锁情况
# 自动资源管理
```
