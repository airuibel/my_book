### 2.3 特征工程

#### 2.3.1什么是特征工程

**专业背景知识和技巧处理数据**，**使得特征能在机器学习算法上发挥更好的作用的过程**。

#### 2.3.2 为什么需要特征工程

数据和特征决定了机器学习的上限，而模型和算法只是逼近这个上限而已。

#### 2.3.3 特征工程包含内容

1. 特征提取

   1. 将任意数据(文本/图像)转成数字特征

2. 特征预处理

   1. 通过转化函数,将特征转化成更加适合模型的数据过程

3. 特征降维

   ​	1. 不损失或少量损失信息,实现维度(特征个数)减少

# 1.6 模型评估

按照**数据集的目标值不同**

目标值-连续  回归

目标值-离散  分类

## 1 分类模型评估

计算准确率

##### 其他评价指标：精确率、召回率、F1-score、AUC指标等

## 2 回归模型评估

##### 均方根误差（Root Mean Squared Error，RMSE）

真实值与预测值之前的信息损失度量

##### 相对平方误差（Relative Squared Error，RSE）、平均绝对误差（Mean Absolute Error，MAE)、相对绝对误差（Relative Absolute Error，RAE)

## 3 拟合

好--拟合效果好

不好-拟合效果差

欠拟合

过拟合

### 3.1 欠拟合

训练集效果不好 测试集效果也不好

没有学习到数据的特征

模型过于简单(说少了)

### 3.2 过拟合

训练集效果好 测试集效果不好

学习训练数据太充分了 

模型过于复杂(说多了)

模型边缘锯齿状
