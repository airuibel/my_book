# 绘制数学函数

```python
# 构造数据
x = np.linspace(-10,10,1000)
y = np.sin(x)
# 实现绘图

# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y)
# 3. 显示图像
plt.show()
```

# 3 柱状图绘制
## 需求-对比每部电影的票房收入

```python
# 数据
movie = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴', '降魔传','追捕','七十七天','密战','狂兽','其它']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
```

```python
# 绘柱状图
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.bar(movie,y,width=0.5, color=['b','r','g','y','c','m','y','k','c','g','b'])
# 修改标题
plt.title("电影票房收入对比",fontsize=20)
# 显示图像
plt.show()
```

```python
# 绘柱状图
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.bar(range(len(movie)),y,width=0.5, color=['b','r','g','y','c','m','y','k','c','g','b'])
# 修改刻度名
plt.xticks(range(len(movie)),movie)
# 修改标题
plt.title("电影票房收入对比",fontsize=20)
# 添加网格
plt.grid(True,alpha=0.8,linestyle='--')
# 显示图像
plt.show()
```
