# 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度

```python
# 构造数据
x = range(60)
y =[random.uniform(15,18) for i in x]
```

```python
# 绘制
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y)
# 3. 显示图像
plt.show()
```

# 实现刻度修改

```python
# 构造数据
x = range(60)
y =[random.uniform(15,18) for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# 绘制
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y)
# 2.1 修改刻度
# plt.yticks(range(0,40,5))
plt.yticks(range(0,40)[::5])
plt.xticks(x[::5],x_ticks[::5])
# 3. 显示图像
plt.show()
```

# 添加描述信息

```python
# 构造数据
x = range(60)
y =[random.uniform(15,18) for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# 绘制
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y)
# 2.1 修改刻度
# plt.yticks(range(0,40,5))
plt.yticks(range(0,40)[::5])
plt.xticks(x[::5],x_ticks[::5])
# 2.2 显示网格
plt.grid(True,linestyle='--',alpha=0.3)
# 2.3 添加描述信息
plt.xlabel("时间",fontsize=18)
plt.ylabel("温度",fontsize=18)
plt.title("一个小时内温度变化",fontsize=20)
# 3. 显示图像
plt.show()
```

# 多次plot

```python
# 构造数据
x = range(60)
y_beijing =[random.uniform(15,18) for i in x]
y_shanghai = [random.uniform(0,3) for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# 绘制
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y_beijing)
plt.plot(x,y_shanghai,linestyle='--',c='r')
# 2.1 修改刻度
# plt.yticks(range(0,40,5))
plt.yticks(range(0,40)[::5])
plt.xticks(x[::5],x_ticks[::5])
# 2.2 显示网格
plt.grid(True,linestyle='--',alpha=0.3)
# 2.3 添加描述信息
plt.xlabel("时间",fontsize=18)
plt.ylabel("温度",fontsize=18)
plt.title("一个小时内温度变化",fontsize=20)
plt.savefig('./test.png')
# 3. 显示图像
plt.show()
```

# 显示图例

```python
# 构造数据
x = range(60)
y_beijing =[random.uniform(15,18) for i in x]
y_shanghai = [random.uniform(0,3) for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# 绘制
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y_beijing,label='北京')
plt.plot(x,y_shanghai,linestyle='--',c='r',label='上海')
# 2.1 修改刻度
# plt.yticks(range(0,40,5))
plt.yticks(range(0,40)[::5])
plt.xticks(x[::5],x_ticks[::5])
# 2.2 显示网格
plt.grid(True,linestyle='--',alpha=0.3)
# 2.3 添加描述信息
plt.xlabel("时间",fontsize=18)
plt.ylabel("温度",fontsize=18)
plt.title("一个小时内温度变化",fontsize=20)
# 2.4 显示图例
plt.legend(loc='upper left')
plt.savefig('./test.png')
# 3. 显示图像
plt.show()
```

```python
# 构造数据
x = range(60)
y_beijing =[random.uniform(15,18) for i in x]
y_shanghai = [random.uniform(0,3) for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# 绘制
# 1. 创建画布
# plt.figure(figsize=(20,10),dpi=100)
fig,axes = plt.subplots(ncols=2,nrows=1,figsize=(20,10),dpi=100)
# 2. 绘制图像
# plt.plot(x,y_beijing,label='北京')
axes[0].plot(x,y_beijing,label='北京')
# plt.plot(x,y_shanghai,linestyle='--',c='r',label='上海')
axes[1].plot(x,y_shanghai,linestyle='--',c='r',label='上海')
# 2.1 修改刻度
# plt.yticks(range(0,40,5))
# plt.yticks(range(0,40)[::5])
# plt.xticks(x[::5],x_ticks[::5])
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_ticks[::5])
axes[0].set_yticks(range(0,40,5))
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_ticks[::5])
axes[1].set_yticks(range(0,40,5))
# 2.2 显示网格
# plt.grid(True,linestyle='--',alpha=0.3)
axes[0].grid(True,linestyle='--',alpha=0.8)
axes[1].grid(True,linestyle='--',alpha=0.8)
# 2.3 添加描述信息
# plt.xlabel("时间",fontsize=18)
# plt.ylabel("温度",fontsize=18)
# plt.title("一个小时内温度变化",fontsize=20)
axes[0].set_xlabel("时间",fontsize=18)
axes[0].set_ylabel("温度",fontsize=18)
axes[0].set_title("一个小时内温度变化",fontsize=18)
axes[1].set_xlabel("时间",fontsize=18)
axes[1].set_ylabel("温度",fontsize=18)
axes[1].set_title("一个小时内温度变化",fontsize=18)
# 2.4 显示图例
# plt.legend(loc='upper left')
axes[0].legend(loc='upper left')
axes[1].legend(loc='upper left')
# 3. 显示图像
plt.show()
```
