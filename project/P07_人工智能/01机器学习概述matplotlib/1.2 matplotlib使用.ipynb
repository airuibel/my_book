# 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度

# 添加自定义x,y刻度


```python
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
# 构造数据
x=range(60)
y=[random.uniform(15,18) for i in x]
x_ticklabel=["11点{}分".format(i) for i in x]
# 绘制图像
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y)
# 2.1 修改刻度
plt.xticks(x[::5],x_ticklabel[::5])
# plt.yticks(range(0,40,5))
plt.yticks(range(40)[::5])
# 3. 显示图像
plt.show()
```

#  在一个坐标系中绘制多个图像
##  多次plot

```python
# 构造数据
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]
x_ticklabel=["11点{}分".format(i) for i in x]
# 绘制图像
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y_shanghai)
plt.plot(x,y_beijing,c='r',linestyle='--')
# 2.1 修改刻度
plt.xticks(x[::5],x_ticklabel[::5])
# plt.yticks(range(0,40,5))
plt.yticks(range(40)[::5])
# 2.2 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)
# 2.3 添加描述信息
plt.xlabel("时间",fontsize=18)
plt.ylabel('温度',fontsize=18)
plt.title('上海11点温度变化',fontsize=20)
# 2.4  图像保存
plt.savefig('./my45.png')
# 3. 显示图像
plt.show()
```

# 显示图例

```python
# 构造数据
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]
x_ticklabel=["11点{}分".format(i) for i in x]
# 绘制图像
# 1. 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 2. 绘制图像
plt.plot(x,y_shanghai,label='上海')
plt.plot(x,y_beijing,c='r',linestyle='--',label='北京')
# 2.1 修改刻度
plt.xticks(x[::5],x_ticklabel[::5])
# plt.yticks(range(0,40,5))
plt.yticks(range(40)[::5])
# 2.2 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)
# 2.3 添加描述信息
plt.xlabel("时间",fontsize=18)
plt.ylabel('温度',fontsize=18)
plt.title('上海11点温度变化',fontsize=20)
# 2.4  图像保存
plt.savefig('./my45.png')
# 2.5 显示图例
plt.legend(loc='upper left')
# 3. 显示图像
plt.show()
```

# 多个坐标系显示— plt.subplots

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(20,10))
```

```python
axes[0][0]
```

```python
# 构造数据
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]
x_ticklabel=["11点{}分".format(i) for i in x]
# 绘制图像
# 1. 创建画布
# plt.figure(figsize=(20,10),dpi=100)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,10),dpi=100)
# 2. 绘制图像
# plt.plot(x,y_shanghai,label='上海')
# plt.plot(x,y_beijing,c='r',linestyle='--',label='北京')
axes[0].plot(x,y_shanghai,label='上海')
axes[1].plot(x,y_beijing,label='北京',c='r')
# 2.1 修改刻度
# plt.xticks(x[::5],x_ticklabel[::5])
# plt.yticks(range(0,40,5))
# plt.yticks(range(40)[::5])
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_ticklabel[::5])
axes[0].set_yticks(range(0,40,5))
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_ticklabel[::5])
axes[1].set_yticks(range(0,40,5))
# 2.2 添加网格显示
# plt.grid(True,linestyle='--',alpha=0.5)
axes[0].grid(True,linestyle='--',alpha=0.5)
axes[1].grid(True,linestyle='--',alpha=0.5)
# 2.3 添加描述信息
# plt.xlabel("时间",fontsize=18)
# plt.ylabel('温度',fontsize=18)
# plt.title('上海11点温度变化',fontsize=20)
axes[0].set_xlabel("时间",fontsize=18)
axes[0].set_ylabel('温度',fontsize=18)
axes[0].set_title("上海温度变化",fontsize=20)
axes[1].set_xlabel("时间",fontsize=18)
axes[1].set_ylabel('温度',fontsize=18)
axes[1].set_title("北京温度变化",fontsize=20)
# 2.4  图像保存
# plt.savefig('./my45.png')
# 2.5 显示图例
# plt.legend(loc='upper left')
axes[0].legend(loc='best')
axes[1].legend(loc='best')
# 3. 显示图像
plt.show()
```
