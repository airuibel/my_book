````python
import pandas as pd
import numpy as np
````

````python
# 数据的基本操作
````

````python
data = pd.read_csv("./data/stock_day.csv")
````

````python
data = data.drop(["ma5","ma10","ma20","v_ma5","v_ma10","v_ma20"], axis=1)
````

````python
data.head(10)
````

````python
# 先列后行
data["open"]["2018-02-27"]
````

````python
# 结合loc和iloc使用索引
# 使用loc:只能指定行列索引的名字
````

````python
# 获取从'2018-02-27':'2018-02-22'，'open'的结果
data.loc["2018-02-27":"2018-02-22","open"]
````

````python
# 使用iloc可以通过索引的下标去获取
# 获取前3天数据,前5列的结果

data.iloc[:3,:5]
````

````python
# 使用ix 组合索引
# 获取行第1天到第4天，['open', 'close', 'high', 'low']这个四个指标的结果
````

````python
data.ix[:4,['open', 'close', 'high', 'low']]
````

````python
# 用loc
data.loc[data.index[:4],['open', 'close', 'high', 'low']]
````

````python
# 用iloc
data.iloc[:4,data.columns.get_indexer(['open', 'close', 'high', 'low'])]
````

````python
# 赋值操作
data["close"] = 1
````

````python
data.head()
````

````python
# 排序
# 使用df.sort_values(by=, ascending=)
# 单个键或者多个键进行排序,
# 参数：
# by：指定排序参考的键
# ascending:默认升序
# ascending=False:降序
# ascending=True:升序
````

````python
# 按照开盘价大小进行排序 , 使用ascending指定按照大小排序
data.sort_values(by='open',ascending=True).head()
````

````python
# 按照多个键进行降序
data.sort_values(by=['open','turnover'],ascending=False).head()
````

````python
# 给索引排序 sort_index
````

````python
data.sort_index().head()
````

````python
data['p_change'].sort_values().head()
````
