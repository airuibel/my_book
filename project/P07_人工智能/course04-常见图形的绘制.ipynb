## Matplotlib能够绘制折线图、散点图、柱状图、直方图、饼图。

### 绘制折线图 plt.plot

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# 构造数据
x = [random.randint(0,10) for i in range(5)]
y = np.linspace(-10,10,5)
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.plot(x,y)
# 显示图像
plt.show()
```

### 散点图 plt.scatter

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# 构造数据
x = np.linspace(0,10,20)
y = (np.e)**x
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.scatter(x,y)
# 显示图像
plt.show()
```

### 绘制柱状图plt.bar

```python
# 构造数据
x = range(10)
y = np.random.uniform(1,10,10)
print(x)
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.bar(x,y,width=0.5,color=['b','r','g','y','c','m','y','k','c','g','b'])
# 显示图像
plt.show()
```

### 直方图plt.hist

```python
# 构造数据
# x = range(10)
data = np.random.randn(100)
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
# data:必选参数，绘图数据
# bins:直方图的长条形数目，可选项，默认为10
# normed:是否将得到的直方图向量归一化，可选项，默认为0，代表不归一化，显示频数。normed=1，表示归一化，显示频率。
# facecolor:长条形的颜色
# edgecolor:长条形边框的颜色
# alpha:透明度
plt.hist(data, bins=10,facecolor="blue", edgecolor="black", alpha=0.7)
# 显示图像
plt.show()
```

### 饼图的绘制 plt.pie(x,lables-,autopct=,colors)

```python
# 构造数据
# 创建画布
plt.figure(figsize=(10,10),dpi=100)
# 绘制图像
plt.pie([1,2,3,4],labels=['java',"c++","javascript","python"],autopct="%.1f%%")
# 显示图像
plt.show()
```

### 柱状图练习


```python
# 构造数据
movie = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴', '降魔传','追捕','七十七天','密战','狂兽','其它']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.bar(movie,y,width=0.5,color=['b','r','g','y','c','m','y','k','c','g','b'])
# 添加标题
plt.title("电影票房数据",fontsize=20)
plt.xlabel("电影名称",fontsize=18)
plt.ylabel("电影票房",fontsize=18)
# 显示网格
plt.grid(linestyle='-.',alpha=0.6)
# 显示图像
plt.show()
```
