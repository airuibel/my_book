# 数据离散化
- 连续属性的离散化就是在连续属性的值域上，将值域划分为若干个离散的区间，最后用不同的符号或整数 值代表落在每个子区间中的属性值。

````python
import pandas as pd
stock = pd.read_csv("./data/stock_day.csv")
stock.head()
````

````python
# pd.qcut(data,q) 对数据进行分组，一般会与value_count搭配使用，统计每组的个数
# series.value._counts() 统计分组次数
p_change = stock['p_change']
# 分组
cut = pd.qcut(p_change,10)
````

````python
# 计算分到每个组数据的个数
cut.value_counts()
````

````python
# 自定义分组
bins = [-100,-7,-5,-3,0,3,5,7,100]
p_counts = pd.cut(p_change,bins)
````

````python
p_counts
````

````python
p_change.value_counts()
````

````python
# 股票涨跌分组数据变成one-hot编码
````

## one-hot 编码
```
pandas.get_dummies(data, prefix=None)

data:array-like, Series, or DataFrame

prefix:分组名字
```

````python
dummies = pd.get_dummies(p_counts,prefix="rise")
````

````python
dummies
````

````python
# 实现表合并
pd.concat([stock,dummies],axis=1)
````
