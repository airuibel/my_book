# 数据运算

```python
data['open']+1
```

```python
data['open'].add(1).sum()
```

```python
data['open']-1
```

```python
data['open'].sub(1)
```

# 逻辑运算

```python
# 例如筛选data["open"] > 23的日期数据
data[data['open']>23]
```

```python
# 获取open 23 到24 之间数据
data[(data['open']>23) & (data['open']<24)]
```

```python
# 获取open<23 和>24 之间数据
data[(data['open']<23) | (data['open']>24)]
```

```python
data.query('open>23 & open<24')
```

```python
# 例如判断'open'是否为23.53和23.85
data[data['open'].isin([23,24])]
```
