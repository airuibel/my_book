```python
import pandas as pd
```

```python
col =pd.DataFrame({'color': ['white','red','green','red','green'], 'object': ['pen','pencil','pencil','ashtray','pen'],'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})
```

```python
col
```

```python
# 统计不同颜色价格均值 DateFrame 写法
col.groupby('color')['price1'].mean()
```

```python
# Series 写法
col['price1'].groupby(col['color']).mean()
```

```python
# 统计不同颜色价格均值 DateFrame 写法
col.groupby('color',as_index=False)['price1'].mean()
```

```python
#  3 星巴克零售店铺数据
starbar = pd.read_csv('./data/starbucks/directory.csv')
```

```python
starbar.head()
```

```python
# 需求统计每个国家都有几家星巴克
count = starbar.groupby('Country').count()['Brand']
```

```python
count.sort_values(ascending=False)[:20].plot(kind='bar',figsize=(20,10))
```

```python
# 统计每个国家每个省市星巴克数量
starbar.groupby(['Country','State/Province']).count()
```
