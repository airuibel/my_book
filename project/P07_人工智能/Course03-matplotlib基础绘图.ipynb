### 准备数据并画出初始折线图

```python
import matplotlib.pyplot as plt
import random
%matplotlib inline
# 生成x坐标
x = range(60)
# 生成y坐标
y = [random.uniform(15,18) for i in x]
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.plot(x,y)
# 显示图像
plt.show()
```

### 添加自定义 x,y刻度

```python
import matplotlib.pyplot as plt
import random
%matplotlib inline

# x坐标
x = range(60)
# 构造x轴刻度标签
x_ticks = ['11点{}分'.format(i) for i in x][::5]
# y坐标
y = [random.uniform(15,18) for i in x]
# 构造 y轴刻度标签
y_ticks = range(40)[::5]
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.plot(x,y)
# x 要显示的刻度值
plt.xticks(x[::5],x_ticks)
# y要显示的刻度值
plt.yticks(y_ticks)
# 显示图像
plt.show()
```

### 设置图形风格

```python
# x坐标
x = range(60)
# 构造x轴刻度标签
x_ticks = ['11点{}分'.format(i) for i in x][::5]
# y坐标
y_shanghai = [random.uniform(15,18) for i in x]
# 构造 y轴刻度标签
y_beijin_ticks = y_shanghai_ticks = range(40)[::5]
y_beijing = [random.uniform(1,4) for i in x]
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制上海温度变化图像
plt.plot(x,y_shanghai,color='r',linestyle='--')
# 绘制北京温度变化图像
plt.plot(x,y_beijing,color='g',linestyle='-.')
# x 要显示的刻度值
plt.xticks(x[::5],x_ticks)
# y要显示的刻度值
plt.yticks(y_ticks)
# x轴描述信息
plt.xlabel("时间",fontsize=18)
# y轴描述信息
plt.ylabel("温度",fontsize=18)
# 表格描述信息
plt.title("中午11点--12点某城市温度变化图",fontsize=21)
# 添加网格显示,linestyle网格线条类型，alpha表示透明度
# - 实线 
plt.grid(True,linestyle="-",alpha=0.8)
# 显示图像
plt.show()
```

### 显示图例

```python
# x坐标
x = range(60)
# 构造x轴刻度标签
x_ticks = ['11点{}分'.format(i) for i in x][::5]
# y坐标
y_shanghai = [random.uniform(15,18) for i in x]
# 构造 y轴刻度标签
y_beijin_ticks = y_shanghai_ticks = range(40)[::5]
y_beijing = [random.uniform(1,4) for i in x]
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制上海温度变化图像
plt.plot(x,y_shanghai,color='r',linestyle='--',label='上海')
# 绘制北京温度变化图像
plt.plot(x,y_beijing,color='g',linestyle='-.',label="北京")
# x 要显示的刻度值
plt.xticks(x[::5],x_ticks)
# y要显示的刻度值
plt.yticks(y_ticks)
# x轴描述信息
plt.xlabel("时间",fontsize=18)
# y轴描述信息
plt.ylabel("温度",fontsize=18)
# 表格描述信息
plt.title("中午11点--12点某城市温度变化图",fontsize=21)
# 添加网格显示,linestyle网格线条类型，alpha表示透明度
# - 实线 
plt.grid(True,linestyle="-",alpha=0.8)

# 显示图例
# plt.legend(loc='best')
plt.legend(loc='upper left')
# 显示图像
plt.show()
```

### 多坐标显示

```python
# x坐标
x = range(60)
# 构造x轴刻度标签
x_ticks = ['11点{}分'.format(i) for i in x][::5]
# y坐标
y_shanghai = [random.uniform(15,18) for i in x]
# 构造 y轴刻度标签
y_beijin_ticks = y_shanghai_ticks = range(40)[::5]
y_beijing = [random.uniform(1,4) for i in x]
# 创建画布
# plt.figure(figsize=(20,10),dpi=100)
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=100)
# 绘制上海温度变化图像
# plt.plot(x,y_shanghai,color='r',linestyle='--',label='上海')
axes[0].plot(x,y_shanghai,color='r',linestyle='--',label='上海')
# 绘制北京温度变化图像
# plt.plot(x,y_beijing,color='g',linestyle='-.',label="北京")
axes[1].plot(x,y_beijing,color='g',linestyle='-.',label="北京")

# x 要显示的刻度值
# plt.xticks(x[::5],x_ticks)
# y要显示的刻度值
# plt.yticks(y_ticks)
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_shanghai_ticks)
axes[0].set_xticklabels(x_ticks)
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_beijin_ticks)
axes[1].set_xticklabels(x_ticks)

# x轴描述信息
# plt.xlabel("时间",fontsize=18)
axes[0].set_xlabel("时间",fontsize=18)
axes[1].set_xlabel("时间",fontsize=18)
# y轴描述信息
# plt.ylabel("温度",fontsize=18)
axes[0].set_ylabel("温度",fontsize=18)
axes[1].set_ylabel("温度",fontsize=18)
# 表格描述信息
# plt.title("中午11点--12点某城市温度变化图",fontsize=21)
axes[0].set_title("中午11点--12点某城市温度变化图",fontsize=21)
axes[1].set_title("中午11点--12点某城市温度变化图",fontsize=21)
# 添加网格显示,linestyle网格线条类型，alpha表示透明度
# - 实线 
# plt.grid(True,linestyle="-",alpha=0.8)
axes[0].grid(True,linestyle="--",alpha=0.8)
axes[1].grid(True,linestyle="-.",alpha=0.8)
# 显示图例
# plt.legend(loc='best')
# plt.legend(loc='upper left')
axes[0].legend(loc='upper left')
axes[1].legend(loc='best')
# 显示图像
plt.show()
```

### 折线图的应用场景

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y = np.sin(x)

# 创建画布
plt.figure(figsize=(20,10),dpi=100)

# 绘制函数图像
plt.plot(x,y)

# 显示网格
plt.grid(True,linestyle=":",alpha=0.5)

# 显示图像
plt.show()
```
