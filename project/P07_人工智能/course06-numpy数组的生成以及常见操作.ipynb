### 生成固定范围的数组

```python
# 创建等差数列
import numpy as np
```

```python
np.linspace(-10,10,50)
```

```python
# 创建等差数列，指定的步长
np.arange(0,10).reshape(2,5)
```

```python
import numpy as np
# 创建等比数列,base为公比
np.logspace(0,3,4,base=2,dtype="int32")
```

```python
np.logspace(0,6-1,6,base=5,dtype="int64")
```

```python
# 生成随机数，正态分布 参数指定数据的形状
np.random.randn(2,3)
```

```python
# 生成标准正态分布,loc为均值，scale为标准差
np.random.normal(loc=0.0,scale=1.0,size=10)
```

```python
# 生成指定形状的标准正态分布
np.random.standard_normal(size=(3,3))
```

```python
# 生成均值为1.75，标准差为1的正态分布数据，100000000个

data = np.random.normal(1.75,1,10000000)
print(data)
```

```python
import matplotlib.pyplot as plt
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.hist(data,1000)
# 显示图像
plt.show()
```

```python
# 创建符合正态分布的4只股票5天的涨跌幅数据
stock_change = np.random.normal(0,1,(4,5))
print(stock_change)
```

```python
# 随机生成4支股票1周的交易日涨幅数据
np.random.randn(4,5)
```

## 均匀分布

```python
# np.rand返回(0,1]之间的数
np.random.rand(5,4)
```

```python
# np.random.uniform(low,high.size) 从一个均匀分布中随机采样，左闭由开

np.random.uniform(10,20,size=(2,3))
```

```python
# 均匀分布图像的生成
# 构造数据
x = np.random.rand(10000000)
print(x)
# 创建画布
plt.figure(figsize=(20,10),dpi=100)
# 绘制图像
plt.hist(x,bins=1000)
# 显示图像
plt.show()
```
