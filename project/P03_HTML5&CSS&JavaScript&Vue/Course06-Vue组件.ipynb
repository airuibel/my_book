# Vue组件语法
```
组件(Component)是Vue.js最强大的功能之一
组件可以封装重用代码
组件都是 Vue 的实例，可接受相同的选项对象 ，并提供相同的生命周期钩子
示例演示
新建目录demo4_1，使用pycharm打开这个目录
在demo4_1目录下，新建目录js，拷贝vue-2.5.16.js到js下
在pycharm编辑器中设置：File->Settings->Languages&Frameworks->JavaScript->选择语言版本为ES6，点击OK按钮
```

## 注册
- 注册组件主要有两种方式：
    - 全局注册
    - 局部注册
- 组件支持嵌套使用

## 全局
- 定义全局组件语法：
- 注意：定义在Vue实例前

```html
<script>
    Vue.component('组件名称', {
        template: `封装的html代码`,
    });
</script>
```

在html中调用语法
注意：写在Vue实例的html内部
<组件名称></组件名称>
示例：创建quanju.html文件，代码如下
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>vue组件</title>
    <script src="js/vue-2.5.16.js"></script>
</head>
<body>
<div id="app">
    <vue_h3></vue_h3>
</div>

<script>
    Vue.component('vue_h3', {
        template: ` <div>
            <h1>hello title</h1>
            <span>hello content</span>
            </div>`,
    });
    vue_app = new Vue({
        el: '#app',
    });
</script>
</body>
</html>
```

## 局部
- 定义局部组件语法：
  - 注意：定义在Vue实例前

```html
<script>
    var 组件名称 = {
        template: `封装的html代码`
    }
</script>
```
注册：在vue实例中使用components属性
```html
<script>
    var vue_app = new Vue({
        el: '#app',
        components: {
            组件名称,
        }
    })
</script>
```
在html中调用语法
注意：写在Vue实例的html内部
<组件名称></组件名称>
示例：创建jubu.html文件，代码如下
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Title</title>
    <script src="js/vue-2.5.16.js"></script>
</head>
<body>
<div id="app">
    <vue_h2></vue_h2>
</div>
<script>
    var vue_h2 = {
        template: `<h2>vue h2 title</h2>`
    }
    var vue_app = new Vue({
        el: '#app',
        components: {
            vue_h2,
        }
    })
</script>
</body>
</html>
```

## 嵌套
- 组件中可以使用components属性嵌套其它组件使用
    - 示例：创建qiantao.html文件，代码如下
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Title</title>
    <script src="js/vue-2.5.16.js"></script>
</head>
<body>
<div id="app">
<vue_h2_1></vue_h2_1>
</div>
<script>
    var vue_h2_2 = {
        template: `<h2>vue h2 2</h2>`
    }
    Vue.component('vue_h2_1', {
        template: `<div>
            <h2>vue h2 1</h2>
            <hr>
            <vue_h2_2></vue_h2_2>
            </div>`,
        components: {
            vue_h2_2,
        }
    });
    var vue_app = new Vue({
        el: '#app',
    });
</script>
</body>
</html>
```

# 特殊属性
- 新特性
    * 在组件中大部分属性与前面学过的是一样的，如methods、mounted等属性
- 新特性如下：
    * data
    * props
    * 单个根元素
    * 事件
    
### data
```
在组件中data是一个函数，而不是字典，并且函数必须返回字典
```

```html
        data:function () {
            return {
                '数据键':'数据值',
                ...
            }
        },
```
示例：创建data.html文件，代码如下
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Title</title>
    <script src="js/vue-2.5.16.js"></script>
</head>
<body>
<div id="app">
    <vue_h1></vue_h1>
</div>

<script>
    Vue.component('vue_h1',{
        template:'<h1 @click="say_hello">{{title}}</h1>',
        data:function () {
            return {
                'title':'hello h1'
            }
        },
        methods:{
            say_hello:function () {
                alert('hello h1');
            }
        }
    });
    var vue_app=new Vue({
       el:'#app',
    });
</script>
</body>
</html>
```

### props
- 在组件中通过props属性为组件标签提供属性
        props:['属性名称',...]
- 示例：创建props.html文件，代码如下
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Title</title>
    <script src="js/vue-2.5.16.js"></script>
</head>
<body>
<div id="app">
<vue_h1 title="vue"></vue_h1>
</div>
<script>
    var vue_h1 = {
        template: `<h1>hello {{title}}</h1>`,
        props:['title']
    }
    var vue_app = new Vue({
        el: '#app',
        components: {
            vue_h1,
        }
    });
</script>
</body>
</html>
```

### 事件
- 调用组件时自定义事件：
    <vue_h1 @自定义事件="处理函数"></vue_h1>
    在组件中定义系统事件，调用自定义事件

```html
this.$emit('自定义事件',参数);
```
示例：创建shijian.html文件，代码如下

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Title</title>
    <script src="js/vue-2.5.16.js"></script>
</head>
<body>
<div id="app">
    <vue_h1 @say="say_hello"></vue_h1>
</div>

<script>
    var vue_h1={
        template:`<input type="button" @click="is_upload('ok')" value="事件">`,
        methods:{
            is_upload:function (txt) {
//                alert(txt);
                this.$emit('say',txt);
            }
        }
    }
    var vue_app=new Vue({
       el:'#app',
        components:{
           vue_h1,
        },
        methods:{
           say_hello:function (txt) {
               alert(txt);
           }
        }
    });
</script>
</body>
</html>
```

# 单文件组件
- 单文件组件：将一个组件相关的html、css、js写在一个单独的文件中
- 一个组件具有了结构、表现和行为的完整功能
- 方便组件之间随意组合以及组件的重用文件扩展名为.vue
    * 注意：文件中的js代码要有空格，并且不能以分号结尾
- 文件中的代码主要包括三段

```html
<template>
     定义html代码
</template>

<script>
    export default{
         定义组件的属性
    }
</script>

// scope关键字，表示这些样式是组件局部的，不会影响其他元素
<style scoped>
     定义css样式
</style>
```

### 主要目录
- 文件夹1(src)，开发目录
- 文件夹components：所有.vue单文件组件都在这个目录下
- 文件夹router：单文件组件的路由都在这个目录下
- 文件夹2(static)，静态资源目录，所有的css，js文件放在这个文件夹
- 文件夹3(dist)，项目打包发布文件夹，编译后的项目文件都在这个文件夹中

- 相关目录
    * node_modules目录是node的包目录
    * config是配置目录
    * build是项目打包时依赖的目录


- 主文件index.html
- index.html中引入src文件夹中的main.js
- main.js中导入顶级单文件组件App.vue
- App.vue中通过组件嵌套或者路由来引用components文件夹中的其他单文件组件

# 路由
## 配置

- 所有的路由规则都配置在src/router/index.js文件中
- 配置index、hello组件的路由

```
import index from '@/components/index'
import hello from '@/components/hello'

Vue.use(Router)

export default new Router({
  routes: [
    {
      path: '/',
      name: 'index',
      component: index
    },
    {
      path: '/index',
      name: 'hello',
      component: hello
    }
  ]
})
```

- 运行项目，默认会访问/路由对应的组件
## 使用
- 链接
```html
<router-link to="路由规则">提示文本</router-link>
重定向
this.$router.push({path:'路由地址'})
```
