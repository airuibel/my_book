```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script>
    /*
        while 条件:
            循环体语句
            增量语句
    JS for 和 while 可以直接转换: 
        初始语句;
        while(条件){
            循环体语句
            增量语句
        }
        for(初识语句;循环条件;增量语句){
            循环体语句
        }
    */
    var o2 = [1,2,3,4,5,6,7,8];
    var iCount = 0;
    while(iCount<o2.length){
        console.log(o2[iCount]);
        iCount ++;
    }
    for(var iCount=0;iCount<o2.length;iCount++){
        console.log(o2[iCount]);
    }
	// do-while 先执行语句  再判断条件
    </script>
</head>
<body>
</body>
</html>
```

### 2.14 定时器

​	作用: 指定时间后执行某一个代码

​	按运行次数分类:  分为单次定时器和多次定时器

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script>

        window.onload = function(){
            // function fnXXX(){alert("单次定时器执行了")}
            // setTimeout(fnXXX, 3000);
            // 1 创建单次定时器   setTimeout(代码入口, 延时时间毫秒,入口所需参数...)
            var oTimer1 = setTimeout(function(number1){
                console.log("单次定时器执行了" + number1);    
                }, 
                3000,  // 延时时间
                100    // 入口代码所需的参数
            );
            // 2 创建多次定时器  setInterval() 参数和单次定时器一致
            var oTimer2 = setInterval(function(){
                    // 1获取图片对象
                    var oImg = document.getElementById("img1");
                    // 2 修改 src 属性
                    console.log(oImg.src);
                    if (oImg.src == "file:///Users/PC/Desktop/JS/img/i2.jpg"){
                        oImg.src = "./img/i3.jpg"
                    }else{
                        oImg.src = "./img/i2.jpg"
                    }
                },
                2000   // 每次定时器调用时间的间隔
            )
            function stop1(){
                clearTimeout(oTimer1);// 3停止单次定时器
            }
            function stop2(){
                clearInterval(oTimer2);// 4停止多次定时器
            }
            document.getElementById('b1').onclick = stop1;
            document.getElementById('b2').onclick = stop2;
        }
    </script>
</head>
<body>
    <img id="img1" src="./img/i2.jpg" alt="">
    <button id="b1">点击我停止单次定时器</button>
    <button id="b2">点击我停止多次定时器</button>
</body>
</html>
```
