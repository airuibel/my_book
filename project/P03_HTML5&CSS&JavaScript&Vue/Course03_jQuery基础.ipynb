```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script src="./js/jquery-1.12.4.min.js"></script>
    <script>
    $(function(){
        //jq 代码写在这里
        // 1 id 选择器      #xxxx 根据标签的 id 属性找标签
        $('#div2').css({background: 'red'});
        $('#div1').css({background: 'blue'})

        // 2 标签选择器     根据标签名字找标签
        $('p').css({'background': "pink"})

        // 3 类选择器    根据标签的 class 属性查找标签
        $('.sp').css({background: 'gold'})

        // 4 后代选择器   根据标签的嵌套结构查找后代标签
        // $('ul li span') 
        $('.ul1 li span').css({background: 'yellow'})

        // 5 属性选择器  根据一个标签的属性去查找标签
        // $('img[src]')
        $('img[alt=i2]').css({width: '250px', height: '250px'})
    });
    </script>
</head>
<body>
    <div id="div1">这是第一个区块</div>
    <div id="div2">这是第一个区块</div>
    <p>段落 1</p>
    <p>段落 2</p>

    <span class="sp">特别提醒 1</span>
    <span>特别提醒 2</span>

    <ul class="ul1">
        <li><span>文字 1</span></li>
        <li><span>文字 2</span></li>
        <li><span>文字 3</span></li>
    </ul>
    <img src="./img/i2.jpg" alt="i2">
    <img src="./img/i3.jpg" alt="i3">
    <img src="./img/i4.jpg" alt="i4">
</body>
</html>
```

### 2.1 选择集过滤

​	选择器运行之后产生的结果是一个集合， 选择集。

​	选择集过滤:  把装有很多标签的集合再加上一定的条件进行筛选，满足条件标签就保留在结果集中，不满足就从结果集中移除了。

​	实现:	.eq	.has

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>
    <script src="./js/jquery-1.12.4.min.js"></script>

    <script>
        $(function(){
            // 选择集过滤: 多个标签中再次加条件进行筛选

            // 1 .eq(下标)  筛选出集合中对应下标的标签
            // $('div').css({background: "red"});
            // $('div').eq(0).css({background: "red"});

            // 2 .has(后代标签对应的选择器)   得到还是 div 标签
            // $('div').has('#mytext').css({background: "red"});
            $('div').has('input[type=text]').css({background: "red"});
        });
    </script>
</head>
<body>
    <div>
        这是第一个div
        <input type="password" id="mytext">
    </div>

<div>
    这是第二个div
    <input type="text">
    <input type="button">
</div>
</body>
</html>
```

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script src="./js/jquery-1.12.4.min.js"></script>
    <script>
    $(function(){
        //  原生 js 绑定事件的写法  对象.事件  = function(){} 
        // 1 jq  绑定事件          jq对象.事件(function(){})
        // 2 常用事件 
        /*click() 鼠标单击  blur() 元素失去焦点 focus() 元素获得焦点
        mouseover() 鼠标进入（进入子元素也触发） mouseout() 鼠标离开（离开子元素也触发）
        */
        // 2.1 单击
        $('#b1').click(function(){
            console.log('客官 你点了什么菜')
        });
        //  2.2 获取焦点
        $('#i1').focus(function(){
            console.log('客官 请输入')
        });
        // 2.3 失去焦点
        $('#i1').blur(function(){
            console.log('客官 慢走')
        });
        // 2.4 鼠标进入
        $('ul li').mouseover(function(){
            // 区分具体是哪个标签发生了事件  this 是一个 js 对象   
            // $(this)就是发生事件的对象
            $(this).css({background: 'red'})
        })
        // 2.5 鼠标移出
        $('ul li').mouseout(function(){
            // 区分具体是哪个标签发生了事件  this 是一个 js 对象   $(this)
            $(this).css({background: 'pink'})
        })
    });
    </script>
</head>
<body>
    <ul>
        <li>元素 1</li>
        <li>元素 2</li>
        <li>元素 3</li>
        <li>元素 4</li>
        <li>元素 5</li>
    </ul>
    <input id="i1" type="text">
    <button id="b1">点我呀</button>
</body>
</html>
```

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script src="./js/jquery-1.12.4.min.js"></script>
    <script>
    $(function(){
        // 普通的事件绑定问题: 如果新增标签不具备同级标签的功能
        // 事件代理: 把事件绑定到父标签 让父标签在子标签触发事件能够调用功能
        // 当用户点击按钮时 在 ul 中增加一对 li
        $('#b1').click(function(){
            $('#ul1').append('<li>元素 </li>')
        })
        // $('ul li').mouseover(function(){
        //     // 区分具体是哪个标签发生了事件  this 是一个 js 对象   
        //     // $(this)就是发生事件的对象
        //     $(this).css({background: 'red'})
        // })
        // 事件代理写法:  父标签.delegate('子标签名','事件名',事件执行的代码)
        $('#ul1').delegate('li','mouseover', function(){
            $(this).css({background: 'red'})
        })
    });
    </script>
</head>
<body>
    <ul id="ul1">
        <li>元素 1</li>
        <li>元素 2</li>
        <li>元素 3</li>
        <li>元素 4</li>
        <li>元素 5</li>
    </ul>
    <button id="b1">点我呀</button>
</body>
</html>
```

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script>
        // JSON是开发过程中非常常用一种数据存储格式   JS对象 和 JSON 进行相互转化
        // 1 创建 JS 对象的第一种方法
        var oMan = new Object();
        // 手动给对象添加一个属性
        oMan.address = "上海";
        oMan.name = "三毛";
        oMan.age = 38;
        oMan.getAge = function(){
            return this.age;
        }

        // 2 对象.属性 获取对象中的属性值
        alert(oMan.name + oMan.address + oMan.getAge())
        
        // 3 建 JS 对象的第二种方法
        var oMan = {name:"小红", age:17, getAge: function(){return this.age;}};
        alert(oMan.name + oMan.age + oMan.getAge());
    </script>
</head>
<body>
    
</body>
</html>
```

>   可以和 JS对象相互转化   stringify    parse

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script>
    // 1 JSON 就是一套数据格式<字符串数据>  所有字符串一律都是用双引号
    // 2 分为对象格式 和 数组格式
    // 3 对象格式 {"k1":v1,"k2":v2,...} 
    var sJSON = '{"name":"张三","age":18,"address":"北京"}'
    var sJSON = '{"name":"张三","age":18,"love":["抽烟","喝酒","洗脚"]}'
    
    // 4 数组格式 [v1,v2,v3,v4,....]
    var sSJON2 = '[18,19,20,"张三",{"name":"hehe"}]'

    // 5.1 在 JSON 中  数组和对象和组合使用的
    // 5.2 能够在 JSON 作为 k 存在是字符串
    // 5.3 能够在 JSON 作为 v 存在 是数字、字符串、对象、数组

    // 6. 和 JS 对象相互转化
    // 6.1 JSON字符串变为 JS对象
    var ojs = JSON.parse(sJSON);
    alert(ojs.love[1])

    // 6.2 JS 对象转化为 JSON 字符串
    var oMan = {name:"小红", age:17, getAge: function(){return this.age;}};
    var oJSONData= JSON.stringify(oMan);
    alert(typeof(oJSONData) + oJSONData)
    </script>
</head>
<body>
    
</body>
</html>
```

应用场景:  局部刷新

后台接口说明:

```
后台地址 http://www.honmon.xyz/Poetry.html
支持方法 GET和 POST
要求参数 name    参数只能是"feng"
返回数据格式  JSON
o =  {"code": 200, "result": [
{"authors": "罗隐", "title": "蜂", "content": "不论平地与山尖，无限风光尽被占。|采得百花成蜜后，为谁辛苦为谁甜。"}
], "message": "成功"}
从结果中取出需要的数据   o.result[0].content
```

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script src="./js/jquery-1.12.4.min.js"></script>
    <script>
        $(function(){
            // jq code
            // $.ajax({k:v,k1:v2})  函数参数就是 js 对象
            $.ajax({
                // 1 后台接口地址
                url: "http://www.honmon.xyz/Poetry.html",
                // 2 请求方法
                type: "GET",
                // 3 请求参数 是一个 JS 对象  {k:v,k:v}
                data:{name:"feng"},
                // 4 后台返回的数据类型
                dataType: "JSON",
                // 5 成功请求时执行的函数代码   函数参数的意义: 响应的数据，状态
                // ajax 在能够理解数据格式的情况下回自动会对数据进行转化为 JS对象
                success:function(response, status){
                    alert("成功了 " + response.result[0].content)
                    $('#p1').html(response.result[0].content);
                },
                // 6 失败请求时执行的函数代码 
                error:function(){
                    alert("失败了")
                }, 
                // 7 是否异步  本次请求是否需要等上一次响应返回  等同步的，不等就是异步
                async: true
            });
        });
    </script>
</head>
<body>
        <h2>蜂</h2>
        <p id="p1"></p>
    <h2>长恨歌</h2>
        <p id="p2">
        汉皇重色思倾国，御宇多年求不得。
		....
        在天愿作比翼鸟，在地愿为连理枝。
        天长地久有时尽，此恨绵绵无绝期。
    </p>
</body>
</html>
```

ajax 简写形式

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Document</title>

    <script src="./js/jquery-1.12.4.min.js"></script>
    <script>
        $(function(){
            // jq code  函数参数是按位置传参 默认异步
            // $.get(
            $.post(
                // 1 后台接口地址
                "http://www.honmon.xyz/Poetry.html",
                // 2 提交给后台的数据
                {name:"feng"},
                // 3 成功时执行的回调函数
                function(response, status){
                    alert("成功了 " + response.result[0].content)
                    $('#p1').html(response.result[0].content);
                },
                // 4 后台返回的数据类型
                "JSON"
            ).error(function(){/*错误时回调*/});
        });
    </script>
</head>
<body>
    <h2>蜂</h2>
        <p id="p1"></p>
    <h2>长恨歌</h2>
        <p id="p2">
        在天愿作比翼鸟，在地愿为连理枝。

        天长地久有时尽，此恨绵绵无绝期。
    </p>
</body>
</html>
```
