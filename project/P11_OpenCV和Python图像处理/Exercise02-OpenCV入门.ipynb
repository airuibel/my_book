```python
# 动图制作
import os
import imageio
from Const import COMBINE_IMAGE
from PIL import Image, ImageDraw, ImageFont


def create_gif(image_list, gif_name):
    frames = []
    for image_name in image_list:
        print(image_name)
        if image_name.endswith('.png'):
            print(image_name)
            frames.append(imageio.imread(image_name))
    # Save them as frames into a gif
    imageio.mimsave(gif_name, frames, 'GIF', duration=0.3)

    return


def clip_image(image_list, region):
    '''
    clip the image and save to a new image
    '''
    for image_name in image_list:
        img = Image.open(image_name)
        cropImg = img.crop(region)
        # cropImg.save(image_name.split(".")[0]+"_new.png")
        cropImg.save(image_name.split(".")[0] + "_large.png")
        # im = Image.open(image_name.split(".")[0]+"_new.png")
        im = Image.open(image_name.split(".")[0] + "_large.png")
        draw = ImageDraw.Draw(im)
        textsize = 160
        ft = ImageFont.truetype("simhei.ttf", textsize)
        text = os.path.basename(image_name)[0:12]
        w, h = draw.textsize(text)

        draw.text((1100, 120), text, font=ft, fill=(0, 0, 0), align="center")
        # im.save(image_name.split(".")[0]+"_new.png")
        im.save(image_name.split(".")[0] + "_large.png")


def custom_image(image_list, region, outdir=""):
    '''
    对图片进行裁剪，标注之类的
    '''
    for image_name in image_list:
        img = Image.open(image_name)
        cropImg = img.crop(region)
        # cropImg.save(image_name.split(".")[0]+"_new.png")
        cropImg.save(outdir + os.sep + os.path.basename(image_name))
        # im = Image.open(image_name.split(".")[0]+"_new.png")
        im = Image.open(outdir + os.sep + os.path.basename(image_name))
        draw = ImageDraw.Draw(im)
        textsize = 40
        ft = ImageFont.truetype("simhei.ttf", textsize)
        text = os.path.basename(image_name)[0:12]
        w, h = draw.textsize(text)

        draw.text((400, 50), text, (255, 255, 255), font=ft, align="center")
        # im.save(image_name.split(".")[0]+"_new.png")
        im.save(outdir + os.sep + os.path.basename(image_name))


def main():
    # path=r'E:\data_between_win_linux'
    # path=r'E:\dataset\shanghai_cases\等高线\profile_barbs\2\uv'
    # path = r'E:\dataset\shanghai_cases\20190818\阵列图片\三维图片\冰雹\image_edit'
    # path=r'D:\公司文档\雷象科技\case_analysis\changsha\20190818\single_radar/'
    path = COMBINE_IMAGE
    image_list = [path + os.sep + img for img in os.listdir(path)]
    gif_name = r'..\combine.gif'
    # clip_image(image_list,[800,1400,2000,2600])
    # clip_image(image_list,[280,400,3400,3200])
    # custom_image(image_list,[50,250,1100,900],outdir=r'E:\dataset\shanghai_cases\20190818\阵列图片\三维图片\冰雹\image_edit')
    create_gif(image_list, gif_name)


if __name__ == "__main__":
    main()
```
