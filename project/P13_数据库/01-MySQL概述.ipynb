## 3. MySQL



### 3.1 简介

数据库作用: 存储和管理数据

分类:关系型<二维结构>和非关系型

特点：高效、保证数据有效、持久化存储



安装:  linux>sudo apt-get install mysql-server mysql-client

服务操作: linux>sudo service mysql status/start/stop/restart

连接服务器: linux>mysql -uroot -pmysql

​	linux> mysql -uroot -p

退出 mysql:  mysql>exit   / quit/ **ctrl d**



### 3.2 数据类型

对数据的强制约束   age = AAA

	大小     看数据的表示范围  选择尽可能小的数据类型
	有无符号 看数据由没有负数
	定长变长 看数据长度是否固定



### 3.3 约束

1.  主键: primary key唯一确定一行记录 <包含唯一性>

​	对上号

​	int unsigned , id

>   是面向当前系统设计的  比如学号

2.  唯一约束: unique   字段值不允许重复    

​	比如手机号  、身份证

3.  非空约束: not null   不允许为空     

​	比如紧急联系人

4.  默认约束: default   字段值没有设置   设置为默认值

    比如性别默认可以设置为保密

5.  外键约束: foreign key

​	有一个表<`子表`>使用另一个表<`主表`>的数据

`如果将这个来自另一个表的数据定义为外键 那么该字段取值将会受到外键约束的限制(**要求子表中外键字段取值 一定要在主表字段值中存在**)`
