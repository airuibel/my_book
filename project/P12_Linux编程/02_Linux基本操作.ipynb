### 3.2 ls查看目录信息

tree 以树状图显示文件目录结构

ls   默认情况查看当前目录信息

作用: 显示目录信息

格式: ls 选项 文件/目录名

选项:

​	-a 显示目录的所有all文件信息(以.开始的文件默认隐藏)

​	-l  以一行line信息一个文件的详细信息

```shell
# 文件详细信息说明
d rwxr-xr-x   2   root root  4096 11月 28  2018 bin
文件类型 权限 硬链接数 用户名 组名 大小 修改时间       文件名
```

```shell
ls -a
ls -l
ls -a -l  #  大多数命令的选项可以连在一起写 不需分前后顺序
ls -l -a
ls -la
ls -al
ls -alh  # 需要结合 l 一起使用
ls -h    # 单独用没有意义
ls -l hello.py  # 查看指定文件的详细信息
ls -l py27      # 查看指定目录的详细信息
ls -l py28
ls -l /
```



### 3.7 rm 删除

作用:  删除文件或者目录

格式: rm 选项 文件名 ...

选项: 

​	-r 递归删除目录下及其子目录

​	-f 忽略不存在的提示

>   rm -rf /   慎用  

​	-i  交互式(需要用户确认删除)

​	-d 删除空目录 (  rm -d 目录 = rmdir 目录)

```shell
rm -i haha.py# 需要用户确认是否删除文件
rm -f 4.txt  # 如果文件存在就删除4.txt 不存在不会提示
rm 4.txt     # 如果文件存在就删除4.txt 不存在会提示
rm -d 222    # 要求目录为空  否则不能删除
```

rm 文件名

rm -r 目录名

### 3.8 cp

作用: 拷贝copy文件或者目录

形式: cp 选项 源路径/文件名  目的路径

常用选项:

​	-r  递归拷贝目录及其下子目录

​	-i  需要用户确认是否覆盖同名文件<不加 i默认是覆盖的>

​	-v 显示拷贝的文件路径信息

​	-a 保留源文件的属性<权限 时间>

```shell
python@ubuntu:~/Desktop/Linux/code$ cp -riav day01/ ../
'day01/1.py' -> '../day01/1.py'

cp -a ../mNetAssist-release-amd64.deb .
-rw-r--r-- 1 python python 139148 7月  19  2017 mNetAssist-release-amd64.deb
```

### 3.9 mv

作用: 移动、重命名  文件或者目录

形式:mv 选项 源路径/源文件  目的路径/新文件

常用选项:

​	-i 交互式(需要用户确认是否覆盖)

​	-v 显示文件的路径信息

```shell
# 1 移动my.txt文件移动用户主目录
mv my.txt ~
# 2 移动一个目录到用户主目录
mv 11 ~
# 3 重名了一个文件
mv 1.py hello.py
# 4 重命名目录  windows 大小不区分   Linux 区分大小写
mv code CODE
# 5 移动文件并且改名
mv mNetAssist-release-amd64.deb CODE/my.deb

# 6 交互式移动文件 如果目的路径下有源文件 则提示是否覆盖
python@ubuntu:~/Desktop/Linux$ mv -i a.txt CODE/my.deb
mv：是否覆盖'CODE/my.deb'？ n

# 7 显示移动的文件路径信息
python@ubuntu:~/Desktop/Linux$ mv -iv a.txt CODE/my.deb
mv：是否覆盖'CODE/my.deb'？ yes
'a.txt' -> 'CODE/my.deb'
```

### 3.12 输出重定向

概念: 将原本输出到终端运行的结果 转而输出到文件中

形式: 命令  符号 文件名

种类: > >>

```shell
# > 覆盖输出重定向  将命令结果直接覆盖写入到文件中去  w
ls > a.txt 

# >> 追加输出重定向  命令结果追加写到文件中去       a
ls -ahl >> a.txt
```
