### 1.pycharm远程开发

* 第一步：上传代码
  * 1、pycharm中选择tools---选择部署---选择加号----sftp----给服务起名---填入主机和用户名密码
  * 2、配置部署的文件路径，建议和本地路径的文件名一样。
  * 3、上传部署的文件
* 第二步：选择解释器
  * 1、添加add---->选择ssh解释器---->选择已经存在的配置----->选择toutiao-web----->点击create(copy)----->选择远程的python解释器home/python/.virtualenvs/toutiao/bin/python------>同步文件夹
  * 2、代码运行，如果再win系统里访问，需要修改host=0.0.0.0，通过真实ip访问。
