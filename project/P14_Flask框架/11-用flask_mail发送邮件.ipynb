```python
import time
from datetime import timedelta, datetime
from flask import Flask
from flask_mail import Mail, Message

app = Flask(__name__)

# 配置邮件：服务器/端口/传输安全协议/邮箱名/密码
app.config.update(
    DEBUG=True,
    MAIL_SERVER='smtp.qq.com',
    MAIL_PROT=465,
    MAIL_USE_TLS=True,
    MAIL_USERNAME='xxx@foxmail.com', # 发送邮件的邮箱
    MAIL_PASSWORD='***',         # 邮件的鉴权码
)

# 实例化邮件
mail = Mail(app)

@app.route("/send_email/")
def aps():
    # sender 发送方，recipients 接收方列表
    msg = Message("This is a test", sender='mrtluh@foxmail.com', recipients=[
        "mrtluh@163.com"])
    # 邮件内容
    msg.body = "你好世界"
    with app.app_context():
        mail.send(msg)
    print("send mail success")
    now_time = time.localtime()
    # 标准化时间
    now_time = time.strftime("%Y-%m-%d %H:%M:%S", now_time)
    # 将时间转换为datetime的时间对象
    now_time = datetime.strptime(now_time, "%Y-%m-%d %H:%M:%S")
    end_time = now_time + timedelta(minutes=30)
    return "</h1>邮件发送成功！！！{}后再发送</h1>".format(end_time)


if __name__ == '__main__':
    app.run(host="127.0.0.1")
```

````python
# 保存发送邮件定时任务到数据库

from apscheduler.schedulers.background import BackgroundScheduler
# 导入任务存储模块
from apscheduler.jobstores.sqlalchemy import SQLAlchemyJobStore
from pymysql import install_as_MySQLdb
install_as_MySQLdb()

import time
from datetime import timedelta,datetime
from flask import Flask
# Mail用来发送邮件，Message用来构造邮件内容、标题等信息；
from flask_mail import Mail, Message

app = Flask(__name__)
# 配置邮件：服务器／端口／传输层安全协议／邮箱名／密码
app.config.update(
    DEBUG = True,
    MAIL_SERVER='smtp.qq.com',
    MAIL_PROT=465,
    MAIL_USE_TLS = True,
    MAIL_USERNAME = '371673381@qq.com',
    MAIL_PASSWORD = 'bmnid',
)

mail = Mail(app)
# 指定任务存储的mysql数据库
db_url = 'mysql://root:mysql@localhost/demo_cron'
job_stores = {
    'default': SQLAlchemyJobStore(url=db_url)
}
bs = BackgroundScheduler(jobstores=job_stores)

# 发送邮件的功能函数
def index():
    # sender 发送方，recipients 接收方列表
    msg = Message("This is a test ",sender='371673381@qq.com', recipients=['shengjun@itcast.cn'])
    # #邮件内容
    msg.body = "Flask test mail"
    # 发送邮件
    with app.app_context():
        mail.send(msg)
    print('send mail success')
    return "Send　Succeed"

@app.route('/')
def aps():
    t = time.localtime()
    # 2019-10-22 10:55:01标准日期和时间格式，2020-01-01
    now_time_str = '%d-%02d-%02d %d:%d:%d' % (t.tm_year, t.tm_mon, t.tm_mday, t.tm_hour, t.tm_min, t.tm_sec)
    now_time = datetime.strptime(now_time_str, '%Y-%m-%d %H:%M:%S')
    end_time = now_time + timedelta(minutes=30)
    bs.add_job(index,'interval',start_date=now_time,end_date=end_time,seconds=20)

    bs.start()
    return 'send mail success'


if __name__ == "__main__":
    app.run()
````
