```python
# 距离5号线大屯路东站843米 随时看房 集中供暖
# 地铁口  距离
```

```python
import re
re.search(pattern='距离(.+号线)(.+站)([0-9]+)米',string = '距离5号线大屯路东站843米 随时看房 集中供暖').group(3)
```

```python
def fin_sub_station(x):
    try:
        return re.search(pattern='距离(.+号线)(.+站)([0-9]+)米',string = x).group(2)
    except:
        return None
lj_data.loc[:,'sub_station'] = lj_data['extra_info'].apply(fin_sub_station)
```

```python
tmp_sub = lj_data[['sub_station','view_num']].groupby('sub_station',as_index = False).count()
```

```python
tmp_sub.sort_values('view_num',ascending = False)
```

```python
def has_sub_station(x):
    return 1 if "距离" in x else 0

lj_data.loc[:,'has_sub_station'] = lj_data['extra_info'].apply(has_sub_station)

lj_data[['has_sub_station','price']].groupby('has_sub_station').agg('mean')
```

```python
def cal_sub_station_distance(x):
    try:
        return int(re.search(pattern="(.+号线)(.+站)([0-9]+)米", string=x).group(3))
    except:
        return None

lj_data.loc[:,'distance'] = lj_data['extra_info'].apply(cal_sub_station_distance)

lj_data['distance'].mean()
```

```python
def get_floor(x):
    if '低楼层' in x:
        return '低楼层'
    elif '中楼层' in x:
        return '中楼层'
    else:
        return '高楼层'

lj_data.loc[:,'floor'] = lj_data['floor'].apply(get_floor)
```
