#### Excel 简介

- excel函数
  - countif  数一下满足条件的记录有几条
    - =Countif(判断的范围，“要满足的条件“)
  - countifs 数一下满足多个条件的记录有几条
    - =Countif(判断的范围1，“要满足的条件1”，判断的范围2，“要满足的条件2”)
  - sumif  对某一列所有满足条件的记录求和
    - =SUMIF(B2:B8,">3000",B2:B8)
    - =SUMIF(判断数据的记录范围,"条件",求和的范围)
  - sumifs  对某一列所有满足多个条件的记录求和
    - =SUMIFS(求和范围,判断的范围1，“要满足的条件1”，判断的范围2，“要满足的条件2”)
  - rank   排序
    - =RANK(指定要排序的单元格,排序的范围）
    - 列锁定  在（用相同公式填充）拖动的时候 把内容锁定可以加上$符号
  - 行列锁定

- 逻辑函数
  - and  与
    - =AND(C2>=500,C2<1000)
  - or     或
    - =OR(C2<500,C2>1000)
  - not   非
    - =NOT(C2<500)
  - if      判断条件，满足条件返回传入的第一个值，否则返回第二个值
    - =IF(C2<500,"便宜","不便宜")

- Match & Index
  - 1、使用MATCH函数返回某个标签所在的位置（组合成行、列坐标）
  - 使用index函数检索得到二维表中的数值。
  - 构建下拉列表
    - 点数据 ， 数据验证
    - 允许-》序列 ，来源 到表格里圈选

- vlookup,hlookup

  - 指定一个值，到一个数据表去查，值作为Key，表中key对应的value就是要找到内容

  - vlookup  要查找的表是纵向的

    =VLOOKUP(D2,L:M,2,FALSE)*P$1

    =VLOOKUP(要查找的K,要查找的数据表,value所在表中的列,精确匹配是0，否则是1)*P$1

  - hlookup 要查找的表是横向的

    ==HLOOKUP(要查找的K,要查找的数据表,value所在表中的行,精确匹配是0，否则是1)*P$1

- indirect
  - 先有一个表，用来查找
  - 通过indirect 传入单元格的位置，返回该单元格中的值
  - =INDIRECT(G11)   包含表中单元格位置的单元格位置，不需要引号
  - =INDIRECT("A2")  直接传入单元格的位置 需要用引号
  - 通过indirect可以创建二级索引
- 字符串处理函数
  - left,right 从左边/右边截取字符串
  - mid 从中间截取
    - =MID(D1,8,4)  参数1 要处理的字符串，参数2 从哪里开始截取，参数3 要截取几个
  - find 查找字符串
    - =FIND("传智",D1) 参数1 要查什么，参数2 去哪里查
  - len 字符串长度
  - substitute 字符串替换
    - =SUBSTITUTE(D1,"北京","深圳")  在哪里替换，老字符串，新字符串
  - concatenate 和 & 字符串拼接 
    - =CONCATENATE(C1,"在",C3,"学",C2)
    - =C1&"在"&C3
  - upper 和 lower 大写 小写转换
