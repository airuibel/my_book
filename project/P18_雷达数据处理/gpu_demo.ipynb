````python
import numpy as np
radar_n = np.load("./2019081013370000nradarn.npz")
data = radar_n["arr_0"]
bordar_line = np.zeros_like(data,dtype=np.int8)
for k in range(200):
    d1 = data[k]
    data0 = np.where(d1==1,0,d1)
    # 求出外圈
    for i in range(1279):
        if i in range(1,1278):
            for j in range(1292):
                if j in range(1,1291):
                    if (data0[i,j]-data0[i-1,j]) == -2:
                        data0[i,j] = -1
                    if (data0[i,j]-data0[i+1,j]) == -2:
                        data0[i,j] = -1
                    if (data0[i,j]-data0[i,j-1]) == -2:
                        data0[i,j] = -1
                    if (data0[i,j]-data0[i,j+1]) == -2:
                        data0[i,j] = -1
    # 求出内圈
    for i in range(1279):
        if i in range(1,1278):
            for j in range(1292):
                if j in range(1,1291):
                    if (data0[i,j]-data0[i-1,j]) == 3:
                        data0[i,j] = 1
                    if (data0[i,j]-data0[i+1,j]) == 3:
                        data0[i,j] = 1
                    if (data0[i,j]-data0[i,j-1]) == 3:
                        data0[i,j] = 1
                    if (data0[i,j]-data0[i,j+1]) == 3:
                        data0[i,j] = 1
    bordar_line[k,:,:] = np.where(np.logical_or(data0==1,data0==-1),data0,0)
    print("=========第%s层==========")
np.savez("border_line.npz",bordar_line)
````
