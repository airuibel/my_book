### 2.2 链接文件ln⭐️⭐️⭐️

作用:

​	类似于 windows 系统中快捷方式

​	给源文件创建一个`链接link` 文件，用户通过链接文件间接使用/修改 源文件。

分类:  软链接<更常用>  硬链接

分类:  软链接<更常用>  硬链接ln

​	软  链接文件 symbol

​		ln -s 源文件 软链接文件  

​		特点: 大小和源文件没有关系，不存数据 存的是源文件的路径。详细信息第一个字母是 l 。如果源文件删除了  软链接失效。

​		链接文件 symbol

```shell
ln -s data.txt datas.txt
# 如果源文件和链接文件不再一个目录  这样创建软链接是失败的 
ln -s data.txt /home/python/Desktop/datas.txt
# 需要给源文件加绝对路径
ln -s /home/python/Linux/data.txt /home/python/Desktop/datas.txt
# 软链接可以对目录
 ln -s /home/python/Linux/test/ /home/python/Desktop/tests
```

​	硬

​		特点: 如果源文件删除了  硬链接继续使用

​		ln 源文件 硬链接文件

​	硬链接本质上`就是 源文件数据的别名 ； 删除源文件时只是删除一个名字,其他名字仍然可以访问文件数据`。硬连接数其实就是文件数据的别名数量。

>   Ctrl c 终止一个程序   
>
>    ll =ls -al 缩写 部分发行版提供

### 2.3  文本搜索⭐️⭐️⭐️

作用: 一个数据去一个文件中搜索

格式: grep 选项 数据 文件名

常用选项: 

​	忽略大小写    -i

​	显示匹配行号 -n

​	条件取反       -v

```shell
python@ubuntu:~/Desktop/Linux$ grep 'itcast.cn' grep.txt
www.itcast.cn
itcast.cn
python@ubuntu:~/Desktop/Linux$ grep -i 'itcast.cn' grep.txt
www.itcast.cn
WWW.ITCAST.CN
itcast.cn
ITCAST.CN
python@ubuntu:~/Desktop/Linux$ grep -in 'itcast.cn' grep.txt
1:www.itcast.cn
2:WWW.ITCAST.CN
6:itcast.cn
7:ITCAST.CN
python@ubuntu:~/Desktop/Linux$ grep -inv 'itcast.cn' grep.txt
3:ia
```



```shell
# ^匹配行数据以 xxx 开始的
python@ubuntu:~/Desktop/Linux$ grep 'www' grep.txt 
python@ubuntu:~/Desktop/Linux$ grep '^www' grep.txt 

# $匹配行数据接 结尾的
python@ubuntu:~/Desktop/Linux$ grep 'cn' grep.txt 
python@ubuntu:~/Desktop/Linux$ grep 'cn$' grep.txt

# .匹配非换行符的任意一个字符
grep 'i.c' grep.txt
```

### 2.4 文件查找⭐️⭐️⭐️

​	作用: find 能够根据文件的名称特征 会在路径及其所有子目录查找文件

​	格式: find 路径 -name '名字特征'

```shell
# 查找当前系统所有的 hello.py 文件
find / -name 'hello.py'

# 结合通配符使用 *任意个任意字符 ?一个任意字符
find / -name '*.py'
find / -name '?.py'

# 1 搜索系统下所有的 py 文件 
# 2 获取 其中含有/home/python的数据
# 3 不要 其中以/home/python/.开始的数据
sudo find / -name '*.py' | grep '^/home/python/'| grep -v '^/home/python/\.'
```

>   sudo 命令       # 以系统管理员的权限运行命令

### 2.5 压缩和解压⭐️⭐️⭐️

​	tar 命令

格式: tar 选项 压缩文件名 [需要被压缩的文件名...]

常用: -z gz 格式解/压缩  -c 创建归档文件  -v 显示详细过程 -f 指定归档文件名

​	-j bz2 格式

```shell
# 1 将很多文件压缩成 tar.gz /.tgz
tar -zcvf 1.tar.gz *.txt
tar -zcvf test1.tar.gz a.txt *.sql sl_3.03-17build1_amd64.deb 
# 2 解压 gz 文件
tar -zxvf test1.tar.gz

# 3 将很多文件压缩成 tar.bz2
tar -jcvf test2.tar.bz2 a.txt *.sql sl_3.03-17build1_amd64.deb 

# 4 解压 bz2 文件
tar -jxvf test2.tar.bz2

# 5 -C 解压到指定路径
tar -jxvf test2.tar.bz2 -C ~/Desktop
```

>   tar 命令的选项 可以不要-   tar jxvf test2.tar.bz2

### 2.7 权限chmod⭐️⭐️

```shell
# ll  areas.sql
-rwxrwxr-x  1 python python 208268 11月 29  2018 areas.sql*

# r 读  w 写  x 执行权限
# change mode
# user u     group g     other o        all a
# 用户python  同组用户权限  其他用户
# rwx            rwx         r-x

# 两种方式可以修改文件  
# 1 字母法
chmod [a/u/g/o] [+-=] [rwx] 文件名
chmod a-x areas.sql
chmod u-x,o-r,g=rw areas.sql

# 2 数字法<高手必会>  更直接
# chmod 数字数字数字 文件
# 每个数字都代表一组权限 rwx  三个数字代表三组权限
# r4     w2    x1
# 用一个数字 将每个权限之和累计
# 7 rwx      6rw-    0---  4r-- 2-w-
chmod 751 file1  # rwxr-x--x
chmod 755 file1  # rwxr-xr-x
chmod 644 file1  # rw-r--r--
chmod 666 file1  # rw-rw-rw- 
```

### 2.9 用户管理⭐️⭐️

>   用户管理需要 sudo 或者 当前用户是 root
>
>   sudo -s 切换到 root 用户  输入当前管理员的密码

```shell
#1 创建一个用户
sudo useradd laowang -m  # 创建用户 创建用户主目录
sudo useradd laotie      # 不会创建主目录 只创建用户
python@ubuntu:/home$ sudo useradd py28 -m
python@ubuntu:/home$ ls
laowang  py23  py24  py25  py28  python  test3
python@ubuntu:/home$ sudo useradd -m py29 -g python
 
#2 查看用户信息   cat /etc/passwd
python@ubuntu:/home$ id py28
uid=1005(py28) gid=1005(py28) 组=1005(py28)
python@ubuntu:/home$ id py29
uid=1006(py29) gid=1000(python) 组=1000(python)

#3 切换用户身份
sudo passwd py28 # 先给 py28用户设置一个密码 才能切换
su - py28   # 切换到 py28 用户并且自动切换到这个用户的主目录
su py28     # 切换到 py28 用户但不切换当前路径 
su root     # 输入 root 用户密码

#4 删除一个用户delete
sudo userdel laowang      # 删除 laowang  但不删主目录
sudo userdel -r laowang1  # 删除 laowang1 用户/主目录创建一个用户
```

### 2.10 用户组管理⭐️⭐️

>   sudo 或者 root 用户

```shell
# 1 创建一个用户组
python@ubuntu:/home$ sudo groupadd heihei
# 2 查看一个用户组的信息
python@ubuntu:/home$ cat /etc/group
# 3 修改用户所属的组  
# 将 py28 的所属组修改为 heihei
python@ubuntu:/home$ sudo usermod py28 -g heihei
# 将 py28 的所属附加组修改为 sudo 
python@ubuntu:/home$ sudo usermod py28 -G sudo
python@ubuntu:/home$ id py28
uid=1005(py28) gid=1006(heihei) 组=1006(heihei)

sudo usermod py28 -G sudo
id py28 # uid=1005(py28) gid=1006(heihei) 组=1006(heihei),27(sudo)
# 4 附加组的操作
sudo gpasswd -d py28 sudo #正在将用户“py28”从“sudo”组中删除
id py28#uid=1005(py28) gid=1006(heihei) 组=1006(heihei)

sudo gpasswd -a py28 sudo#正在将用户“py28”加入到“sudo”组中
id py28#uid=1005(py28) gid=1006(heihei)组=1006(heihei),27(sudo)
# 5 删除一个用户组  用户组中必须没有用户了 否则出错
sudo groupdel heihei
```

>   一个用户只能加入一个 group 用户组，但是可以加入多个附加组
>
>   最常用的附加组是 sudo《如果你需要一个用户能使用 sudo 命令这个用户需要加入到 sudo 附加组中》
