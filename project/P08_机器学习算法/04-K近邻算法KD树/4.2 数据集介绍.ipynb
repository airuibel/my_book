# 鸢尾花数据图例

```python
iris_d = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
iris_d['target'] = iris.target
```

```python
def iris_plot(data,col1,col2):
    sns.lmplot(col1,col2,data,hue='target',fit_reg=False)
    plt.title('鸢尾花种类分布')
    plt.show()
```

```python
iris_plot(iris_d,'sepal width (cm)','petal width (cm)')
```
