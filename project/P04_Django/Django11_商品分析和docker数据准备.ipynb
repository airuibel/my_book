## 2.fastDFS 分布式文件存储
*   1.普通的上传下载: -- 图片--->本地路径--->读取二进制流-->post--服务器(mysql 远程地址  url)--磁盘二进制流
*   2. 分布式fastDFS 上传下载流程
*   client 客户端
*   tracker 追踪服务器---负载均衡 服务
*   storeage --存储 文件的
*   上传流程:  clinet--上传请求-->tracker(1.服务器空, 2. 所有已上传的文件对应的地址)-->闲着服务器的ip:port-->client
*   clinet---ip:port-->storage-->(1.存二进制流 2.生成文件绝对路径--url) --->client

## 5.镜像 image
* 1.查询当前docker里面所有的镜像 sudo docker images
*                            sudo docker image ls
* 2.安装镜像 : 1. 连外网--- sudo docker image pull 镜像的名字:版本号
*             2. 本地安装-- sudo docker image load -i xxx本地包
* 3.删除镜像:  sudo docker image rm 镜像的名字:版本号/镜像ID


## 6.容器 container
* 1.查询正在运行容器 sudo docker container ls
* 2.查询所有容器 sudo docker container ls -a
* 3.运行镜像  sudo docker run --name= "" -dit 镜像名字:版本号
* 4.停止  sudo docker container stop
* 5.删除  sudo docker container rm
* 
## 7.容器制作 成本地镜像
* 1.将容器 制作成 镜像 sudo docker commit 容器名字 镜像的名字
* 2. 将镜像导出  sudo docker save -o 文件名字.tar  镜像名字
* 3. 验证  sudo docker image load -i 文件名字.tar 


## 8.使用docker 安装fastdfs client tracker storage
*    1. sudo docker image load -i 本地fastdfs.tar
*    2. 运行stracker容器 sudo docker run -dit --name tracker --network=host -v /var/fdfs/tracker:/var/fdfs delron/fastdfs tracker
*    3.运行 storage容器 sudo docker run -dti --name storage --network=host -e TRACKER_SERVER=**真实IP**:22122 -v /var/fdfs/storage:/var/fdfs delron/fastdfs storage
