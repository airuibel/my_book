## 复习回顾
#### web开发流程 4个
*   1. 接收参数
*   2. 解析参数 (参数校验必须写的 (1.前端可能校验错误 2.ajax,其他途径访问的请求, postman,fiddler,charles))
*   3. 交互数据库orm 处理业务逻辑 
*   4. 返回响应结果 (1,前端需要数据格式 JSON (分离)  2. render(request,'xx.html',context)(前后端不分离))

####请求方式:4个 GET(查询) POST(新增) DELETE(删除) PUT(修改)
#### 传参的四种方式:4个
*   1.路径 ----- 正则?P<username>---def get(self,request,username)
*   2.查询参数--- request.GET
*   3.请求体
    *    3.1 form request.POST ==>queryDict
    *    3.2 非表单json request.body ==>bytes
*   4.请求头 request.META
#### vue 基本操作:
*   1.隐藏和显示 v-show
*   2.input select textarea ---v-model
*   3.操作属性 v-bind: 简写 :
*   4. 操作事件 v-on 简写 @
*   5. 大量的重复标签: v-for
*   6. 条件判断 v-if v-else-if
*   7.操作数据 {{  }} --- [[ ]]
*   8. vue的生命周期 beforecreate created beforeMount Mounted updated destroy

#### django数据库查询:
*   增 save() create()
*   删 delete()
*   改 update(); save()
*   查 get filter,all(),exlcude()
*   关联查询 1:n ; 多模型小写_set == related_name; n:1 外键属性
*   关联过滤查询 filter(1:n 多模型小写__多模型的属性__符号=值)
