## 创建项目



### 创建项目的环境

- 先进入Django的虚拟环境，然后输入以下命令

      -  django-admin startproject 项目名称
      


## 配置数据库
- Django中默认用SQlite数据
- 通过settings.py中的DataBases选项配置数据库

### 配置Mysq

    * python3.x安装的是pymysql
    * 在__inti__.py文件中写入两行代码
      - import pymysql
      - pymysql.install_as_MySqldb()
    * 然后创建数据库
      - mysql -u root -p
      - create database student；
    * 配置数据库
   - DATABASES = {
        'default': {
            #'ENGINE': 'django.db.backends.sqlite3',
            'ENGINE': 'django.db.backends.mysql',
            #'NAME': os.path.join(BASE_DIR, 'db.sqlite3'),
            'NAME':'student',
            'USER':'服务器账户名',
            'PASSWORD':'服务器密码',
            'HOST':'localhost',
            'PORT':'3306',
        }


#### 创建应用



- 一个项目中可以创建多个应用
- python manage.py startapp (应用名)

   *   admin.py 站点配置
   *   models.py 模型
   *   views.py 视图
