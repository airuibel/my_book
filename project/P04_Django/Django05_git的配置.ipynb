## 3. [重点]Git的操作流程 PPT 

- 1.网管  运维 技术总监--自己--远程服务器-创建仓库 文件夹(.git)
- 2.克隆 git clone 远程仓库到本地电脑
- 3.写代码-->暂存区--git commit --本地仓库区--git push--远程仓库
- 4.同步别人的代码 -git pull

## 4. [体会] 工作区-暂存区-本地仓库区---远程仓库

- 工作区: 写代码的地方(增删改查)
- 暂存区: 暂时存储一段时间的代码区域
- 本地仓库区: 在本地电脑 记录版本 备份代码的
- 远程仓库: 多人合并代码的地方, 记录版本 备份代码的


## 5. Git单人本地仓库操作
- 1.安装 ubuntun  sudo apt-get install git
- 2.测试是否成功 git 
- 3.创建本地Git仓库  git init
- 4.配置个人信息  git config user.name
                git config user.email
- 查看隐藏文件的 快捷键 contrl+h

- 5.操作代码
- 写代码-->暂存区--git commit --本地仓库区
-  工作区--暂存区 git add xx.py
-  git commit -am "不能提交文件"

- 6.查看历史记录
         1. git log
         2. git reflog  #  样式不一样 ; 记录回退的版本历史记录, 删除记录

- 7.回退版本
        - git reset --hard  HEAD^(粘贴版本号)

- 8.撤销修改:只能撤销工作区和暂存区的代码; 仓库区的回退版本
- 1.工作区: git checkout xx.py
                contrl+z 
        - 2.暂存区-->工作区--撤销写错代码
                - 1. git reset HEAD login.py
                - 2. git checkout xx.py


- 9.对比版本: git diff 
        - 1. 仓库区 和 工作区 对比  git diff HEAD login.py
        - 2. 仓库区自己对比  git diff HEAD  HEAD^


- 10.删除文件
    - 1. touch          rm
    - 2. git add        git rm
    - 3. git commit     git commit
