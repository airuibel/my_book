# 1. Could not connect to Redis at 127.0.0.1:6379: Connection refused
    *   1.原因 -没有连接成功redis的服务器
    *   2.解决办法: 1.服务是否开启 2.端口号是否是对的 3.IP是否是对的
    *   查进程 ps aux|grep redis-server
    *   杀进程 sudo kill -9 进程ID号

    *   ***如果 kill 不掉*** 使用  /etc/init.d/redis-server stop

# 2. Fatal error, can't open config file 'rediss.conf'
    *   1.原因: 当前路径 没有这个文件
    *   1.检查文件名字 2.检查路径

# 3. ERR invalid DB index 下标越界 redis
    *   1.redis数据只有16个  下标 0-15

# 4. MISCONF Redis is configured to save RDB snapshots, but is currently not able to persist on disk. Commands that may modify the data set are disabled. Please check Redis logs for details about the error.
    *   1.原因：强制关闭Redis快照导致不能持久化
    *   2.解决:客户端运行 config set stop-writes-on-bgsave-error no

# 13. 403 CSRF verification failed. Request aborted.
